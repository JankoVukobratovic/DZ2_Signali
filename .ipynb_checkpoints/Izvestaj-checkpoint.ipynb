# Zadatak 2: FDM
## a)
Niskopropusni filteri služe da "odseku" frekfencije više od neke granice.
Oni od signala koji ima frekfencije od ($-\infty$, $+\infty$) (tj. ako posmatramo samo realan deo signala od (0, $+\infty$)
propuštaju samo ($0$, $f_{cutoff}$).
Ovaj element je koristan zato što prilikom sabiranja signala sprečava da visoke frekfencije manje amplitude
prolaze u opseg drugog signala.

Element modulisanja sa $cos(2\pi f_{carrier} t)$ menja opseg frekfencija signala za tu frekfenciju. Tj, pomera opseg.
Ovo je korisno da se u istom trenutku propuštaju informacije oba signala (jedan u sopstvenom opsegu frekfencija, jedan u pomerenom opsegu)

Propusnik opsega ima svrhu sličnu kao Niskopropusni filter, ali opseg frekfencija izlaznog signala može biti $(a,b)
